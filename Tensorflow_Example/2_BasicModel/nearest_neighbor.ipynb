```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets("MNIST_data/", one_hot = True)
```

```python
X_train, Y_train = mnist.train.next_batch(10000)
X_test, Y_test = mnist.train.next_batch(500)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
```

```python
# define placeholder
x_train = tf.placeholder(tf.float32, [None, 784])
x_test = tf.placeholder(tf.float32, [784])
print(x_train, x_test)
```

```python
# define calcul:
distance = tf.reduce_sum(tf.pow(tf.add(x_train, tf.negative(x_test)), 2), axis = 1)
predict = tf.arg_min(distance, dimension = 0)

accuracy = 0
```

```python
with tf.Session() as sess:
    accuracy = 0
    for i in range(X_test.shape[0]):
        predict_index = sess.run(predict, feed_dict = {x_train : X_train, x_test : X_test[i, :]})
        # you can use tf.equal, see it in logistic regression
        #correct_res  = tf.arg_max(Y_test[i], dimension = 0)                # wrong , it will return a tensorflow object: 
        #predict_res = tf.arg_max(Y_train[predict_index], dimension = 0)    # Tensor("ArgMax_1058:0", shape=(), dtype=int64)
        correct_res  = np.argmax(Y_test[i])
        predict_res = np.argmax(Y_train[predict_index])
        if(correct_res == predict_res):
            print("The {} example, Correct, the result = {}".format(i+1, predict_res))
            accuracy += 1
        else:
            print("The {} example,Wrong, the correct result = {}, the predict result = {}".format(i+1, correct_res, predict_res))
            plt.imshow(np.reshape(X_test[i], (28, 28)))
            plt.show()
            plt.imshow(np.reshape(X_train[predict_index], (28, 28)))
            plt.show()
    accuracy = accuracy / X_test.shape[0]

print("accuracy is = {}".format(accuracy))
```
